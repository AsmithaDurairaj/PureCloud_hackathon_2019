```python
import datetime
def compute_resolution_duration_in_hrs(open,close):
    open_date= datetime.datetime.strptime(open, '%Y-%m-%d %H:%M:%S')
    close_date= datetime.datetime.strptime(close, '%Y-%m-%d %H:%M:%S')
    diff = close_date - open_date
    days, seconds = diff.days, diff.seconds
    hours =float( days * 24 + seconds / 3600)
    return hours
```

```python
dataset2['Resolution_duration_hrs'] = dataset2.apply(lambda row : compute_resolution_duration_in_hrs(row['Date/Time Opened'],row['Date/Time Closed']), axis = 1)
```

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

corpus = []    
for i in range(0, 31920):
    review = re.sub('[^a-zA-Z]', ' ', str(dataset2['Product'][i])) #[^a-zA-Z] - this denotes that - "dont remove the characters a-zA-Z.
    #And always removing a character will stick together the adjacent words -to prevent this we give space as second argument.This will replace removed chars by space
    review = review.split()    
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review) # to reverse the list to a string.before this step review is a list of words [].So we are joining all the data in the list with space in between the each word in the list
    corpus.append(review)
dataset2['Product']=corpus
print(corpus)
```
